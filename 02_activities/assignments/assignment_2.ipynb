  Given the `root` of a binary tree, return all root to leaf paths in any order.

#### Starter Code for Question 2

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val = 0, left = None, right = None):
#         self.val = val
#         self.left = left
#         self.right = right

#import list
from typing import List

#define the node
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.left = left
        self.right = right
        self.val = val

def bt_path(root: TreeNode) -> List[List[int]]:
    #create an empty list to hold everything 
    results  = []
    def leafy (node, path):
        if node is None:
            return
        #line to entend the path 
        path = path + [node.val]
        #if there is no child left and right, save it . 
        if not node.left and not node.right:
            results.append (path)
            return
        # if not a leaf , go deeper into children?
        leafy(node.left, path)
        leafy(node.right, path)
# start the root  with an empty path
    leafy(root, [])
    return results 
```
